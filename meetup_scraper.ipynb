```python
from bs4 import BeautifulSoup
import re as re
import pandas as pd
import requests

#Decipher the websites url to point to the page we want to scrape
#Meet up uses query parameteres to point to a page.
#Use requests to load the website HTML code

london_url = "https://www.meetup.com/find/?slug=infused-in-the-community&eventId=293633728&location=gb--17--London&source=EVENTS&eventType=inPerson&keywords=data"
manchester_url = "https://www.meetup.com/find/?slug=infused-in-the-community&eventId=293633728&location=gb--18--Manchester&source=EVENTS&eventType=inPerson&keywords=data"
london_page = requests.get(london_url)
manchester_page = requests.get(manchester_url)
#print(page.text)

#Parse the HTML content by creating a BeautifulSoup object
london_soup = BeautifulSoup(london_page.content, "html.parser")
manchester_soup = BeautifulSoup(manchester_page.content, "html.parser")

#Extract the HTML object using the elemts ID - Where events are located on the page. (Using devloper tool on web browser)
london_results = london_soup.find(id="main")
manchester_results = manchester_soup.find(id="main")
#print(results.prettify())

#Find events by HTML class name
london_event_elements = london_results.find_all("div", class_="p-0 bg-clip-padding bg-cover bg-transparent relative h-full flex bg-white z-0 break-words transition-shadow duration-300 w-full flex-row justify-start py-4 border-t border-gray3 md:pt-4 md:pb-5")
manchester_event_elements = manchester_results.find_all("div", class_="p-0 bg-clip-padding bg-cover bg-transparent relative h-full flex bg-white z-0 break-words transition-shadow duration-300 w-full flex-row justify-start py-4 border-t border-gray3 md:pt-4 md:pb-5")


for event_element in london_event_elements:
    print(event_element, end="\n"*2)

for event_element in manchester_event_elements:
  print(event_element, end ="\n"*2)

#Using child elements extract information required from event postings
for event_element in london_event_elements:
    l_title_element = event_element.find("h2", class_="text-gray7 font-medium text-base pt-0 pb-1 line-clamp-3") #event title
    l_company_element = event_element.find("p", class_="hidden md:line-clamp-1 text-gray6") #group name & location
    l_time_element = event_element.find("time") #date and time

    print(l_title_element.text)
    print(l_company_element.text.replace('Group name:', ''))
    print(l_time_element.text)
    print()

for event_element in manchester_event_elements:
    m_title_element = event_element.find("h2", class_="text-gray7 font-medium text-base pt-0 pb-1 line-clamp-3") #event title
    m_company_element = event_element.find("p", class_="hidden md:line-clamp-1 text-gray6") #group name & location
    m_time_element = event_element.find("time") #date and time

    print(m_title_element.text)
    print(m_company_element.text.replace('Group name:', ''))
    print(m_time_element.text)
    print()

#Insert scraping results into a dataframe

#create empty lists
title = []
company = []
time = []

for event_element in london_event_elements:
    l_title_element = event_element.find("h2", class_="text-gray7 font-medium text-base pt-0 pb-1 line-clamp-3") #title
    l_company_element = event_element.find("p", class_="hidden md:line-clamp-1 text-gray6") #group name & location
    l_time_element = event_element.find("time") #date and time

for event_element in manchester_event_elements:
    m_title_element = event_element.find("h2", class_="text-gray7 font-medium text-base pt-0 pb-1 line-clamp-3") #title
    m_company_element = event_element.find("p", class_="hidden md:line-clamp-1 text-gray6") #group name & location
    m_time_element = event_element.find("time") #date and time

#Extract text from HTML elements
    l_title_element = l_title_element.text
    l_company_element = l_company_element.text.replace('Group name:', '')
    l_time_element = l_time_element.text

    m_title_element = m_title_element.text
    m_company_element = m_company_element.text.replace('Group name:', '')
    m_time_element = m_time_element.text

#Append output to lists
    title.append(l_title_element)
    company.append(l_company_element)
    time.append(l_time_element)

    title.append(m_title_element)
    company.append(m_company_element)
    time.append(m_time_element)

l_links = []
m_links = []

for tag in london_soup.find_all('a', {'href': True, 'class' : 'w-full inline cursor-pointer relative hover:no-underline'}):
  l_link_element = tag['href']

  l_links.append(l_link_element)

for tag in manchester_soup.find_all('a', {'href': True, 'class' : 'w-full inline cursor-pointer relative hover:no-underline'}):
  m_link_element = tag['href']

  m_links.append(m_link_element)

#DataFrames

df1 = pd.DataFrame(title, columns=["Event name"])
df2 = pd.DataFrame(company, columns=["Company and Location"])
df3 = pd.DataFrame(time, columns=["Date and Time"])
df4 = pd.DataFrame(l_links, columns=["Event Link L"])
df5 = pd.DataFrame(m_links, columns=["Event Link M"])
#Joining dataframes to join outputs together
final_results = df1.join(df2).join(df3).join(df4).join(df5)
final_results

#Export to csv
from google.colab import drive
drive.mount('/content/drive')

filename = 'events.csv'
final_results.to_csv('/content/drive/MyDrive/' + filename)
```
